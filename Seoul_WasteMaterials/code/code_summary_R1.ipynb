```r
trash <-read.csv('data/newtest.csv', sep=',')
```

```r
summary(trash)
```

```r
install.packages('corrplot')
```

```r
library(corrplot)
```

```r
head(trash)
```

```r
pairs(trash, panel=panel.smooth)
```

```r
cor <- cor(trash[1:6])
corrplot(cor, method='number')
```

```r
hist(trash$배출량, probability=T)
lines(density(trash$배출량), col='red')
```

```r
hist(trash$음식점, probability=T)
lines(density(trash$음식점), col='red')
```

```r
plot(trash$배출량, trash$음식점, tmain='주류판매 음식점 수에 따른 배출량',  xlab='음식점', ylab='배출량(톤/일)', col='red', pch=15, axes=F)
text(trash$배출량, trash$음식점, cex = 0.7, adj=c(0, -0.5), lab=trash$배출량)
axis(side=1, at=seq(1,7,0.1))
axis(side=2, at=seq(0,3,0.1))
box()
grid()
```

```r
plot(trash$배출량, trash$음식점, col='blue')
abline(lm(trash$음식점~trash$배출량), col='purple')
```

```r
plot(trash$배출량, trash$주민부담률, col='black', pch=6,  xlab='배출량', ylab='처리비용 주민부담률')
abline(lm(trash$주민부담률~trash$배출량), col='purple')
```

```r
plot(trash$배출량, trash$재정자립도, col='blue', xlab='배출량', ylab='처리비용 재정자립도', pch=5)
abline(lm(trash$재정자립도~trash$배출량), col='purple')
```
